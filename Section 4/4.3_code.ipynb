## Using Seaborn for Data Visualization

Seaborn is a data visualization library based on matplotlib.  It offers customized themes and high-level interfaces. Seaborn also works well with pandas dataframes.

```python
#  https://seaborn.pydata.org/
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example warehouse data set
rate = pd.read_csv('make_rate.csv')
rate.head()
```

### Line and Scatter plots

```python
# Plot of relationship
sns.relplot(x="months", y="products", data=rate, kind="line", color='blue')
```

```python
# Plot of linear relationship
sns.relplot(x="months", y="process", data=rate, kind="line", color='coral')
```

```python
# scatterplot of relationship, do not need to specify kind
sns.relplot(x="products", y="months", data=rate)
```

```python
# scatterplot of relationship
sns.relplot(x="products", y="months", hue="rate", data=rate)
```

### Category Plots

```python
# Let's take a look at the outcome variable, frequency counts
sns.catplot(x="rate", kind="count", data=rate);
```

```python
# scatterplot
sns.catplot(x='months', y='products',data=rate)
```

```python
# Violin plots, shows the probability density of the data at different values
sns.catplot(x='months', y='products',kind="violin",data=rate)
```

```python
# Category plot with kind as "boxen", box plot
sns.catplot(x='months', y='products', kind="boxen",data=rate)
```

```python
sns.catplot(x = 'months', y = 'process', kind="boxen", data=rate)
```

### Distribution Plots

```python
from scipy import stats
# loc specifies the mean
# size is number of observations
# scale is the standard deviation
c = np.random.normal(loc=0, size=100, scale=2)
sns.distplot(c)
```

```python
# a single variable distribution plot
sns.distplot(rate['products'], color='brown')
```

```python
# a single variable distribution plot
sns.distplot(rate['process'], color='purple')
```

```python
# joint plot function to show a bivariate distribution
# kde plots a kernel density estimate in the margins and 
# converts the interior into a shaded countour plot
with sns.axes_style("white"):
    sns.jointplot(x=rate['products'], y=rate['process'], kind="kde",color="b");
```

```python
# joint plot function to show a bivariate distribution
# kde plots a kernel density estimate in the margins and 
# converts the interior into a shaded countour plot
with sns.axes_style("white"):
    sns.jointplot(x=rate['products'], y=rate['months'], kind="kde",color="g");
```

```python
# side-by-side facet grids, split according outcome variable 2 groups
x = sns.FacetGrid(rate, col="rate")
x.map(plt.hist, "process")
```

```python
# side-by-side facet grids, split according outcome variable 2 groups
x = sns.FacetGrid(rate, col="rate")
x.map(plt.hist, "products", color='pink')
```

```python
# another way for side-by-side grids, using pair grid
sns.set(style='darkgrid')
# Pair of variables to compare
dat = sns.PairGrid(rate)
dat.map(plt.scatter, color='orange')
```

```python
# pair (scatter) plots
# histograms shown on diagonal
features = rate[['products', 'process', 'months']]
sns.pairplot(features)
```

```python
# Boxplot
sns.set(style="white", color_codes=True)
sns.boxplot(x='months', y='products',data=rate)
```

```python
# Boxplot
sns.set(style="white", color_codes=True)
sns.boxplot(x='months', y='process',data=rate)
```
